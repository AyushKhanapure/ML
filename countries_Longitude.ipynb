```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('countries_coordinates.csv')
```

```python
X = data[['latitude', 'longitude']]
```

```python
num_clusters = 3
```

```python
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
```

```python
kmeans.fit(X)
```

```python
data['cluster'] = kmeans.labels_
```

```python
print(data)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['longitude'], data['latitude'], c=data['cluster'], cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 0], color='red', s=200, alpha=0.75, marker='X')
plt.title('K-means Clustering of Countries by Latitude and Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid()
plt.show()
```
