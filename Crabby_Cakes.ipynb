```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
years_worked = np.array([1, 2, 3, 4, 5, 6])
cakes_made = np.array([6500, 7805, 10835, 11230, 15870, 16387])
```

```python
plt.scatter(years_worked, cakes_made, color='blue', label='Cakes made')
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(years_worked, cakes_made)
```

```python
line_of_best_fit = slope * years_worked + intercept
```

```python
plt.plot(years_worked, line_of_best_fit, color='red', label=f'Line of Best Fit: y = {slope:.2f}x + {intercept:.2f}')
```

```python
plt.xlabel('Years Worked')
plt.ylabel('Cakes Made')
plt.title('Crabby Cakes Made Over Years Worked')
plt.legend()
plt.show()
```

```python
print(f"Equation of the line of best fit: y = {slope:.2f}x + {intercept:.2f}")
```

```python
print(f"Correlation Coefficient (r): {r_value:.2f}")
```

```python
years = 10
predicted_cakes = slope * years + intercept
print(f"Predicted Crabby Cakes after {years} years: {predicted_cakes:.2f}")
```
